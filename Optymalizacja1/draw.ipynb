```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('fibonacci.csv')
```

```python
plt.figure(figsize=(10, 6))

plt.plot(df.index[:180], df['Va'][:180], label='Va', marker='o')

plt.plot(df.index[:180], df['Vb'][:180], label='Vb', marker='x')

plt.xlabel('N')
plt.ylabel('Objentość')
plt.title('Wykres Va i Vb Fibonacci')
plt.legend()

plt.show()
```
